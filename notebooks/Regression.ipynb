```python
sorted_by_args = np.argsort(X)
outlier = sorted_by_args[-2]
X_without_outlier = X[sorted_by_args[:-2]]
Y_without_outlier = Y[sorted_by_args[:-2]]
```

```python
def plot_data(X: np.ndarray, Y: np.ndarray):
    plt.scatter(X, Y, color="tab:blue", label="real data")
    plt.xlabel("x - GDP", fontsize=14)
    plt.ylabel("y - happiness", fontsize=14)

start, stop = X.min(), X.max()

def visualise_compare_LSS_method(X: np.ndarray, T: np.ndarray,
                                 color="tab:orange", label="estimated function"):
    """
    Visualise LSS model on fancy Matplotlib plot.

    :param X: input argument vector
    :param Y: input target vector
    :param T: theta vector with coefficients of ploynomial
    """
    X_test = np.linspace(start=start, stop=stop, num=300)
    Y_pred = lss.generalised_linear_model(X_test, T)
    plt.plot(X_test, Y_pred, color=color, label=label)
```

```python
degrees = [1, 2, 3, 5, 10, 15, 20]
limits = [None, None, None, None, (-5, 20), None, (-500, 500)]

def compute_error(y, y_pred):
    return np.sqrt(np.mean((y - y_pred) ** 2))

n_cols = 2
n_rows = int((len(degrees) + 1)/ 2)

for i, n in enumerate(degrees, start=1):
    plt.figure(figsize=(30, 20))
    plt.subplot(n_rows, n_cols, i)
    T_degree_n = lss.least_squares_solution(X, Y, n)
    T_degree_n_without_outlier = lss.least_squares_solution(X_without_outlier, Y_without_outlier, n)

    plot_data(X, Y)

    if limits[i - 1] is not None:
        plt.ylim(limits[i - 1][0], limits[i - 1][1])

    visualise_compare_LSS_method(X, T_degree_n)
    visualise_compare_LSS_method(X_without_outlier, T_degree_n_without_outlier,
                                 color="tab:pink", label="estimated without outlier")
    plt.legend()
    plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

args_sorted = np.argsort(X)
X_2 = X[args_sorted]
Y_2 = Y[args_sorted]

def fit_polynomial(degree):
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly_features.fit_transform(X_2.reshape(-1, 1))
    reg = LinearRegression()
    reg.fit(X_poly, Y_2)
    y_pred = reg.predict(X_poly)

    return y_pred

errors = [compute_error(Y_2, fit_polynomial(n)) for n in degrees]

plt.figure(figsize=(12, 10))
plt.plot(degrees, errors, marker='.')
# plt.yscale('log')
plt.xlabel('Degree')
plt.ylabel('Error (log-scale)')
_ = plt.title('Error VS Polynomial Degree')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=50, shuffle=False)

class PolyRegression:

    def __init__(self, degrees):
        self.degrees = degrees 
        self._T = None
        # self.reg = LinearRegression()
        # self.poly = PolynomialFeatures(degree=self.degrees, include_bias=False)

    def fit(self, X, Y):
        # X_poly = self.poly.fit_transform(X.reshape(-1, 1))
        # self.reg.fit(X_poly, Y)
        self._T = lss.least_squares_solution(X, Y, self.degrees)

    def predict(self, x):
        return lss.generalised_linear_model(x, self._T)
        # x_poly = self.poly.transform(x.reshape(-1, 1))
        # return self.predict(x_poly)
    
def rmse(y, y_pred):
    return np.sqrt(np.mean((y - y_pred) ** 2))


degrees = np.arange(1, 20)
test_errors = np.empty(len(degrees), dtype=float)
train_errors = np.empty(len(degrees), dtype=float)

for i, n in enumerate(degrees):

    reg = PolyRegression(n)
    reg.fit(X_train, y_train)
    test_errors[i] = rmse(y_test, reg.predict(X_test))
    train_errors[i] = rmse(y_train, reg.predict(X_train))
```
